```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sig
plt.rcParams["text.usetex"] = True
```

```python
ts = np.linspace(0, 1, 100)
T = 1 #period
V_max = 10
Vs = V_max * sig.sawtooth((2 * np.pi / T) *ts + (np.pi / 2), width = 0.5)
```

```python
fig, ax = plt.subplots()
ax.scatter(ts, Vs, s = 5)
#ax.set_xlim(0, T,)

ax.plot([ts[43]-.003, ts[44]+.003], [1.65, 1.65], color='black', linewidth=1)
ax.text(ts[42], .8, r"$\delta V$")
ax.set_xlabel(r"$t$")
ax.set_ylabel(r"$V$")

# ax.set_xticks([0, 5, 10])
# ax.set_xticklabels([r"$0$", r"$5$", r"$10$"])

ax.set_yticks([V_max, -V_max, 0])
ax.set_yticklabels([r"$V_{\rm max}$", r"$V_{\rm min}$", "0"])

ax.tick_params(labelbottom=False) # labelleft=False

ax.grid(True)
fig.savefig("Figures/single_sweep.png")
```

```python
fig, ax = plt.subplots()
ax.scatter(ts, Vs,  s = 5)
ax.scatter(ts, -1 * Vs,  s = 5)
ax.scatter(ts, Vs - (-1*Vs),  s = 5)
ax.set_xlim(0, T)
plt.legend([r"$V_1$", r"$V_2$",r"$V_1 - V_2$"])

ax.set_xlabel(r"$t$")
ax.set_ylabel(r"$V$")
ax.plot([ts[43]-.003, ts[44]+.003], [1.65, 1.65], color='black', linewidth=1)
ax.text(ts[42],.0, r"$\delta V$")
# ax.set_xticks([0, 5, 10])
# ax.set_xticklabels([r"$0$", r"$5$", r"$10$"])

ax.set_yticks([V_max, -V_max, 0])
ax.set_yticklabels([r"$V_{\rm max}$", r"$V_{\rm min}$", "0"])#, r"$V_{1, \rm max} - V_{2, \rm min}$", r"$V_{1, \rm min} - V_{2, \rm max}$

ax.tick_params(labelbottom=False) # labelleft=False

ax.grid(True)
fig.savefig("Figures/double_sweep.png")
```

```python
#plt.plot(v, (c['ne'][1100:1800]), '.')
plt.plot(a['km'], a['Ne/cm-3'])
plt.title('Electron Density Comparison')
plt.ylabel('Electron Density (Ne/cm-3)')
plt.xlabel('Altitude (km)')
 
# Example data
v = np.array(v)  # Altitudes for dataset c
c_ne = np.array(c['ne'][1100:1800])  # Electron density from dataset c
 
a_km = np.array(a['km'])  # Altitudes for dataset a
a_ne = np.array(a['Ne/cm-3'])  # Electron density from dataset a
 
# Interpolate a_ne to match the altitudes in v
a_ne_interp = np.interp(v, a_km, a_ne)
 
# Calculate the average electron density
average_ne = (c_ne + a_ne_interp) / 2
 
# Plot individual datasets
#plt.plot(v, c_ne, '.', label='Dataset c')
plt.plot(a_km, a_ne, label='IRI Data')
 
# Plot the average data
plt.plot(v, average_ne, linestyle='--', color='black', label='Average of UDIP Data')
 
# Add labels and title
plt.title('Electron Density Comparison')
plt.ylabel('Electron Density (Ne/cm-3)')
plt.xlabel('Altitude (km)')
plt.legend()
 
# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Example data
v = np.array(v)  # Altitudes for dataset c
c_te = np.array(c['Te'][1100:1800])  # Electron temperature from dataset c

a_km = np.array(a['km'])  # Altitudes for dataset a
a_te = np.array(a['Te/K'])  # Electron temperature from dataset a

# Interpolate a_te to match the altitudes in v
a_te_interp = np.interp(v, a_km, a_te)

# Calculate the average electron temperature
average_te = (c_te + a_te_interp) / 2

# Calculate percent error between the average and dataset a
percent_error = np.abs((average_te - a_te_interp) / a_te_interp) * 100

# Plot the original data and the average
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
#plt.plot(v, c_te, '.', label='Dataset c')
plt.plot(a_km, a_te, label='IRI Data')
plt.plot(v, average_te, linestyle='--', color='black', label='Average of UDIP Data')
plt.title('Electron Temperature Comparison')
plt.ylabel('Electron Temperature (K)')
plt.xlabel('Altitude (km)')
plt.legend()

plt.tight_layout()
plt.show()
```
